```python
x=np.arange(0,5,1)
```

```python
y1=x+1
y2=x+2
y3=x+3
y4=x+4
```

```python
plt.subplot(1,2,2)
plt.plot(x,y1,'r--',marker='*')
plt.plot(x,y2,'g--',marker='o')
plt.legend(['x','y'])
plt.subplot(1,2,1)
plt.plot(x,y3,'b--',marker='v')
plt.plot(x,y4,'y--',marker='x')
plt.legend(['x','y'])
```
